```python
# 그대로 파이썬으로 정의하면 오버플로 문제가 나타날 수 있음. 
# def softmax(a) :
#     exp_a = np.exp(a)
#     sum_exp_a = np.sum(exp_a)
#     y = exp_a / sum_exp_a
#     return y
```

### 3.5.2 소프트맥스 함수 구현 시 주의점

```python
a = np.array([1010,1000,990])
np.exp(a)/np.sum(np.exp(a)) #오버플로
```

```python
c = np.max(a)
a-c
```

```python
np.exp(a-c)/np.sum(np.exp(a-c))
```

```python
def softmax(a) :
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

### 3.5.3 소프트맥스 함수의 특징
=> 소프트맥스 함수의 출력을 `확률`로 해석할 수 있음.

### 3.6.2 신경망의 추론 처리

입력뉴런 784개, 출력뉴런 10개

```python
def get_data() :
    (x_train,t_train),(x_test,t_test) = load_mnist(normalize=True,flatten=True,one_hot_label=False)
    return x_test,t_test
```

```python
import pickle

def init_network(): #학습된 가중치 매개변수 사용, 딕셔너리 형태
    with open("dataset/sample_weight.pkl",'rb') as f:
        network = pickle.load(f)
    return network
```

```python
def predict(network,x):
    W1,W2,W3 = network['W1'],network['W2'],network['W3']
    b1,b2,b3 = network['b1'],network['b2'],network['b3']
    
    a1 = np.dot(x,W1)+b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1,W2)+b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2,W3)+b3
    y = softmax(a3)
    
    return y
```

```python
x,t = get_data() #testset만
network = init_network()
```

```python
len(x)
```

```python
accuracy_cnt = 0
for i in range(len(x)) : # x : 10000*784
    y = predict(network,x[i]) #softmax 적용한 결과 반환
    p = np.argmax(y)
    if p ==t[i] :
        accuracy_cnt+=1
        
print("Accuracy",float(accuracy_cnt)/len(x))
```

```python
x,t = get_data() #testset만
network = init_network()

batch_size = 100
accuracy_cnt = 0

for i in range(0,len(x),batch_size) :
    x_batch = x[i:i+batch_size]
    y_batch = predict(network,x_batch)
    p = np.argmax(y_batch,axis=1) #0이면 열, 1이면 행, 행기준으로 가장 큰 것
    accuracy_cnt += np.sum(p==t[i:i+batch_size])

print("Accuacy"+str(float(accuracy_cnt)/len(x)))
```
